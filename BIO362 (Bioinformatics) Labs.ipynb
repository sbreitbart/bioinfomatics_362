## Searching for patterns
The re.search() is a function that determines whether or not a pattern appears somewhere in a string. It takes two strings as
arguments:
1. The first argument is the pattern that you want to search for
2. and the second argument is the string that you want to search.

```python
dna = "ATCGCGAATTCAC"
if re.search(r"GAATTC", dna):
 print("Restriction site found!")
```

```python
# if we put the letter r immediately before the opening quotation mark, then any special characters inside the string are ignored and
# used as they were meant to be in the regular expression.

print("With Special Characters:")
print("Rob\tNess\nWas here.\n")
print("Raw Characters:")
print(r"Rob\tNess\nWas here.\n") # r stands for "raw"
```

```python
# if you only want a portion of the string to be interpreted as raw text, you can leave out the r and use the
# backslash "⧵" in the string itself to "escape" the special character immediately after the backslash.
# This will specifically prevent Python from interpreting the special character as anything but ordinary plain text itself.

print("With Special Characters:")
print("\tRob\tNess\nWas here.\n")
print("Escape the tab character but leave the line break:")
print("\tRob\\tNess\nWas here.\n")
```

### alternation

```python
# check for the presence of an AvaII recognition site, which can have two different sequences: GGACC and GGTCC.
# We could use an or conditional:

dna = "ATCGCGGACCGAATTCAC"
if re.search(r"GGACC", dna) or re.search(r"GGTCC", dna):
 print("restriction site found!")
```

```python
# better way, using a RE called "alternation"

dna = "ATCGCGGACCGAATTCAC"
if re.search(r"GG(A|T)CC", dna):
 print("restriction site found!")

dna = "ATCGCGGACCGAATTCAC"
restriction_site = r"GG(A|T)CC" # can save as an object
if re.search(restriction_site, dna):
 print("restriction site found!")
```

### Character groups
The BisI restriction enzyme cuts at an even wider range of motifs. The restriction site is GCNGC , where N represents any base. We can use the same alternation technique to represent this pattern:

`GC(A|T|G|C)GC`

However, a pair of square brackets with a list of characters inside them can represent "any one of these characters". So the pattern

`GC[ATGC]GC`

will match GCAGC, GCTGC, GCGGC and GCCGC.

```python
dna = "ATCGCGGCTGCAATTCAC"
bisI = r"GC[ATGC]GC"
if re.search(bisI, dna):
 print("restriction site found!")
```

### More tricks

```python
# use a period to indicate that that slot can match ANY character

import re
atpp = r"JK.DH"
if re.search(atpp, "JKSDHFAKJHDF;WKEJF;LASKDJF;ALKDFHJKSFA;SDKJF;AWEIRJHSDLKASDF;LKASJDF;LAKSD"):
    print("Pattern found")
else:
    print("pattern not found")
    
    
# use a ^ to indicate when you DON'T want a character included
atpp = r";L[^S]"
if re.search(atpp, "JKSDHFAKJHDF;WKEJF;LASKDJF;ALKDFHJKSFA;SDKJF;AWEIRJHSDLKASDF;LKASJDF;LAKSD"):
    print("Pattern found")
else:
    print("pattern not found")
    

# use hyphens to search for ranges of characters
atpp = r"J[A-F1-9]"
if re.search(atpp, "JKSDHFAKJHDF;WKEJF;LASKDJF;ALKDFHJKSFA;SDKJF;AWEIRJHSDLKASDF;LKASJDF;LAKSD"):
    print("Pattern found")
else:
    print("pattern not found")
    
    
atpp = r";A[P-Q]"
if re.search(atpp, "JKSDHFAKJHDF;WKEJF;LASKDJF;ALKDFHJKSFA;SDKJF;AWEIRJHSDLKASDF;LKASJDF;LAKSD"):
    print("Pattern found")
else:
    print("pattern not found")
```

### Quantifiers

```python
# 0 or 1 times----------------

# A question mark " ? " immediately following a character means that that character is optional – it can match either zero or
# one times. So in the pattern GAT?C , the T is optional, and the pattern will match either GATC or GAC. If we want to apply a
# question mark to more than one character, we can group the characters in parentheses. For example, in the pattern GGG(AAA)?TTT,
# the group of three A 's is optional, so the pattern will match either GGGAAATTT or GGGTTT. Keep in mind that the group inside
# the parentheses is exact - GGG(AAA)?TTT will not match GGGATTT. Finally, notice that we now have two different roles for
# parentheses in regular expressions; they act both to surround alternations, and to surround sections of patterns for use
# with quantifiers.

random_pattern = "EROQQQQIWEURPOQWIEUEROARPQ392ERO4857345IOWEUPROQWIXCB,MZNXCBVMZNXBDHFEROQLAKSJDHFAKLSJDHSDF92378410923781289"
if re.search("EROQ?", random_pattern):
    print("pattern found")
else:
    print("pattern not found")
    

if re.search("EROQ?A", random_pattern):
    print("pattern found")
else:
    print("pattern not found")
    
    
if re.search("ER?A", random_pattern):
    print("pattern found")
else:
    print("pattern not found")
    
    

# 1 or more times: Use a + ------------------------

pattern2 = "oe one onnnnne two three four four"
if re.search("on+e", pattern2):
    print("pattern found")
else:
    print("pattern not found")
    
if re.search("twi+o", pattern2):
    print("pattern found")
else:
    print("pattern not found")
    
    
    
# 0 or more times: Use a * ---------------

pattern2 = "oe one onnnnne two three four four"
if re.search("on*e", pattern2):
    print("pattern found")
else:
    print("pattern not found")
    
    
if re.search("ag*e", pattern2):
    print("pattern found")
else:
    print("pattern not found")
    
    

    
# specific number of repeats: use {} ----------------------------

pattern3 = "sophie soophie sooophie sooooooooooophie"
if re.search("so{1,3}", pattern3):
    print("pattern found")
else:
    print("pattern not found")
    
if re.search("so{5}p", pattern3):
    print("pattern found")
else:
    print("pattern not found")
```
